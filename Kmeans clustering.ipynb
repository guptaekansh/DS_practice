```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import sklearn
import pandas as pd
import seaborn as sb

import sklearn
from sklearn.datasets import load_iris
```

```python
dataset= load_iris()
```

```python


data= dataset.data

df=pd.DataFrame(data)
df['target']=dataset.target

df.head()



kmeans=KMeans(n_clusters=3, random_state=5)
kmeans.fit(data)

kmeans.labels_
color_codes=np.array(['red','green','blue'])



```

```python
fig,ax= plt.subplots(1,2,figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(x=df.iloc[:,0],y=df.iloc[:,1] , c=color_codes[df['target']],s=25)
plt.title('Actual values')
plt.subplot(1,2,2)
plt.scatter(x=df.iloc[:,0],y=df.iloc[:,1] , c=color_codes[kmeans.labels_],s=25)
plt.title('kmeans predicted values')
```

```python
## due to misclassification relableling the clusters of kmeans as per the true values to better visualize the results
relabel=np.choose(kmeans.labels_, [1,0,2]).astype(np.int64)
```

```python
fig,ax= plt.subplots(1,2,figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(x=df.iloc[:,0],y=df.iloc[:,1] , c=color_codes[df['target']],s=25)
plt.title('Actual values')
plt.subplot(1,2,2)
plt.scatter(x=df.iloc[:,0],y=df.iloc[:,1] , c=color_codes[relabel],s=25)
plt.title('kmeans predicted values')
```

```python
from sklearn.metrics import classification_report
print(classification_report(df['target'],relabel))
```
